# **Let's make models - Faisons des mod√®le**

* **Simplest model is a straight line / Le mod√®le le plus simple c'est une droite**

üìà How to fit the most perfect line / comment cr√©er la droite la plus parfaite

![ChessUrl](https://upload.wikimedia.org/wikipedia/commons/e/ed/Least_Squares.gif "chess")

```python
X_data = df['Year']
Y_data = df['Canada']
```

üêç A python function is going to it for us!

```python
a, c = np.polyfit(X_data, Y_data, 1)
fit_equation = lambda x: a * x + c

def plot_fit1(X, Y, f):
    X_fit = np.linspace(min(X), 2050, 1000)
    Y_fit = f(X_fit)

    fig, ax1 = plt.subplots()
    ax1.plot(X_fit, Y_fit, color='r', alpha=0.5, label='Polynomial fit')
    ax1.scatter(X, Y, s=4, color='b', label='Data points')
    ax1.set_title('Lign fit example')
    ax1.legend()
    plt.show()
    
plot_fit1(X_data, Y_data, fit_equation)
```

* **A more complex model would be a curve / Un mod√®le un peu plus complexe serait une courbe**

```python
a, b, c = np.polyfit(X_data, Y_data, 2)
fit_equation = lambda x: a * x ** 2 + b * x + c

def plot_fit(X, Y, f):
    X_fit = np.linspace(min(X), 2050, 1000)
    Y_fit = f(X_fit)

    fig, ax1 = plt.subplots()
    ax1.plot(X_fit, Y_fit, color='r', alpha=0.5, label='Polynomial fit')
    ax1.scatter(X, Y, s=4, color='b', label='Data points')
    ax1.set_title('Curve fit example')
    ax1.legend()
    plt.show()
    
plot_fit(X_data, Y_data, fit_equation)
```
